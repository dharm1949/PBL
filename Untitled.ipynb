```python
import os
import shutil

a = os.getcwd()
b = os.chdir('E:\DHARM')

txt_path = os.path.join(os.getcwd(), "text")
pdf_path = os.path.join(os.getcwd(), "pdf")
jpg_path = os.path.join(os.getcwd(), "jpg")
excel_path = os.path.join(os.getcwd(), "excel")
mp4_path = os.path.join(os.getcwd(), "mp4")
docs_path = os.path.join(os.getcwd(), "docs")
ipynb_path = os.path.join(os.getcwd(), "ipynb")
zip_path = os.path.join(os.getcwd(), "zip files")

if not os.path.isdir(txt_path):
    os.makedirs(txt_path)
    print('text folder created')
   
if not os.path.isdir(pdf_path):
    os.makedirs(pdf_path)
    print('pdf folder created')
   
if not os.path.isdir(jpg_path):
    os.makedirs(jpg_path)
    print('jpg folder created')

if not os.path.isdir(excel_path):
    os.makedirs(excel_path)
    print('excel folder created')
   
if not os.path.isdir(mp4_path):
    os.makedirs(mp4_path)
    print('mp4 folder created')
   
if not os.path.isdir(docs_path):
    os.makedirs(docs_path)
    print('docs folder created')

if not os.path.isdir(ipynb_path):
    os.makedirs(ipynb_path)
    print('ipynb folder created')

if not os.path.isdir(zip_path):
    os.makedirs(zip_path)
    print('zip folder created')

path = 'E:\DHARM'

files = [f for f in os.listdir(path)]


for file in files:
    file_path = os.path.join(os.getcwd(), file)

    if file_path.endswith('.txt')==True:
        print('move file to text folder')
        shutil.move(file_path, txt_path)
       
    if file_path.endswith('.pdf')==True:
        print('move file to pdf folder')
        shutil.move(file_path, pdf_path)
   
    if file_path.endswith('.jpg')==True:
        print('move file to jpg folder')
        shutil.move(file_path, jpg_path)
   
    if file_path.endswith('.excel')==True:
        print('move file to excel folder')
        shutil.move(file_path, excel_path)
   
    if file_path.endswith('.mp4')==True:
        print('move file to mp4 folder')
        shutil.move(file_path, mp4_path)
   
    if file_path.endswith('.docs')==True:
        print('move file to docs folder')
        shutil.move(file_path, docs_path)
   
    if file_path.endswith('.ipynb')==True:
        print('move file to ipynb folder')
        shutil.move(file_path, ipynb_path)
   
    if file_path.endswith('.zip')==True:
        print('move file to zip folder')
        shutil.move(file_path, zip_path)
```
